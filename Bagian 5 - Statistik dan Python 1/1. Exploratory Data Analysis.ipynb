## ECDF ( Emprical Cumulative Distribution Function)

```python
x = np.sort(df['dem_share'])
y = np.arange(1, len(x)+1) / len(x)
plt.plot(x, y, marker='.', linestyle='none')
plt.xlabel('Presentase suara untuk Obama')
plt.ylabel('ECDF')
plt.margins(0.02) # Keeps data off plot edges
```

```python
def ecdf(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n+1) / n
    return x, y
```

```python
data_pa = df[df['state'] == 'PA']['dem_share']
data_oh = df[df['state'] == 'OH']['dem_share']
data_fl = df[df['state'] == 'FL']['dem_share']
```

```python
x_pa, y_pa = ecdf(data_pa)
x_oh, y_oh = ecdf(data_oh)
x_fl, y_fl = ecdf(data_fl)
```

```python
plt.plot(x_pa, y_pa, marker='.', linestyle='none')
plt.plot(x_oh, y_oh, marker='.', linestyle='none')
plt.plot(x_fl, y_fl, marker='.', linestyle='none')
plt.xlabel('Presentase suara untuk Obama')
plt.ylabel('ECDF')
plt.legend(['PA', 'OH', 'FL'])
plt.margins(0.02) # Keeps data off plot edges
```
